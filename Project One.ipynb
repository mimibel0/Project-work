fixed_acidity_mean = dataset['fixed acidity'].mean()
citric_acid_mean = dataset['citric acid'].mean()
chlorides_mean = dataset['chlorides'].mean()
free_sulfur_dioxide_mean = dataset['free sulfur dioxide'].mean()
total_sulfur_dioxide_mean = dataset['total sulfur dioxide'].mean()
density_mean = dataset['density'].mean()
pH_mean = dataset['pH'].mean()
quality_mean = dataset['quality'].mean()

### Reason for replacing null values with the mean of each column
##### If you have a large dataset with many missing values, you can increase the accuracy of your analysis by filling them in with the mean of the appropriate column.This is due to the fact that the mean is a representative number that can provide a good estimate of the missing value.

```python
dataset['fixed acidity'] = dataset['fixed acidity'].fillna(fixed_acidity_mean)
dataset['citric acid'] = dataset['citric acid'].fillna(citric_acid_mean)
dataset['chlorides'] = dataset['chlorides'].fillna(chlorides_mean)
dataset['free sulfur dioxide'] = dataset['free sulfur dioxide'].fillna(free_sulfur_dioxide_mean)
dataset['total sulfur dioxide'] = dataset['total sulfur dioxide'].fillna(total_sulfur_dioxide_mean)
dataset['density'] = dataset['density'].fillna(density_mean)
dataset['pH'] = dataset['pH'].fillna(pH_mean)
dataset['quality'] = dataset['quality'].fillna(quality_mean)
```

```python
fig, axs = plt.subplots(4, 3, figsize=(15, 15))
axs = axs.flatten()
for i, col in enumerate(numerical_cols):
    category_counts = dataset[col].value_counts()
    axs[i].pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%')
    axs[i].set_title(col)
    fig.suptitle('Distribution of Characteristics')
plt.show()
```

```python
fig, axs = plt.subplots(4, 3, figsize=(20, 20))
axs = axs.flatten()
for i, column in enumerate(numerical_cols):
    axs[i].bar(dataset[column].value_counts().index, dataset[column].value_counts().values)
    axs[i].set_title(column)
    fig.suptitle('Distribution of Characteristics')
plt.show()
```
